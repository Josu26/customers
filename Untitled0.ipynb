```python
df = pd.read_csv('https://github.com/Josu26/customers/blob/main/Customers.csv', nrows=10)
print(df.head())
```

```python
# Mostrar los primeros cinco registros de los datos
print(df.head())
```

```python
# Verificar si hay valores faltantes en nuestros datos
print(df.isnull().sum())
```

```python
# Eliminar filas con valores faltantes
df = df.dropna()
```

```python
# Verificar y eliminar duplicados
print(df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Ver las columnas
df.columns
```

```python
# Seleccionar solo las columnas necesarias
df = df[['CustomerID', 'Age', 'Profession']]
```

```python
# Guardar el DataFrame en un archivo CSV
df.to_csv('customers_processed.csv', index=False)
```

```python
# Importamos las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargamos los datos de nuevo
df = pd.read_csv('https://raw.githubusercontent.com/Josu26/Customers/main/Customers.csv')
```

```python
print(df.describe())
```

```python
plt.hist(df['Age'], bins=10)
plt.title('DistribuciÃ³n de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.countplot(x='Profession', data=df)
plt.title('Cantidad de clientes por profesiÃ³n')
plt.xlabel('ProfesiÃ³n')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Annual Income ($)', data=df)
plt.title('Ingresos versus edad')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()
```
